```python
#use scitools for astropy mplt and pandas
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Question 1:
print("\t\t----- Question 2 ------")
data = pd.read_csv("obesity_study.csv")

print("--- 2a --- ")
print("rows:",data.shape[0])
print("cols:",data.shape[1])


print("--- 2b --- ")

# print(data.iloc[0])

print("gender : Qualitative Nominal")
print("age : Quantitative Discrete")
print("height : Quantitative Continuous")
print("weight : Quantitative discrete")
print("family_overweight : Qualitative Nominal")
print("high_caloric_food : Qualitative Nominal")
print("eat_vegetables : Qualitative Ordinal")
print("main_meals_daily :Qualitative Ordinal")
print("food_between_meals :Qualitative Ordinal")
print("smoke : Qualitative Nominal")
print("water_intake_daily : Qualitative Ordinal")
print("monitor_calories : Qualitative Nominal")
print("physical_activity_frequency : Qualitative Ordinal")
print("tech_device_usage : Qualitative Ordinal")
print("alcohol_frequency : Qualitative Ordinal")
print("transportation : Qualitative Nominal")
print("status : Qualitative Ordinal")
# for label in data.columns:
#     # temp = data.loc[:,label]
#     print(label,":")
```

```python
print("--- 2c --- ")
# Create a column called bmi based on the formula described in the article
# equation => weight / (height^2)
bmiData = data["weight"] / (data["height"]**2)

bmiData = data.assign(bmi=bmiData)
display(bmiData)


print("--- 2d --- ")

# temp1 = bmiData[bmiData['status'] == "Normal_Weight"]['weight']

filtered = []
allStatuses = bmiData['status'].unique()
for stat in allStatuses:
    filtered.append(bmiData[bmiData['status'] == stat]['weight'])
print(allStatuses)


plt.figure(figsize=(15, 6))
plt.boxplot(filtered, labels=allStatuses)
plt.show()

# print(statusCol)
# print(weightCol)

# plt.boxplot(x = weightCol, labels = statusCol)
# plt.show()
```

```python
print("--- 2e --- ")

transport = bmiData['transportation']
transport = transport.apply(lambda x: x == "Walking" or x == "Bike")
# print(transport)

trueCount = transport.sum()
counts = [trueCount, len(transport) - trueCount]
plt.pie(counts,labels=["True","False"])
plt.show()

newData = bmiData.assign(physical_transport=transport)
# newData
```

```python
print("--- 2f --- ")
filtered = []
allStatuses = newData['physical_transport'].unique()
for stat in allStatuses:
    filtered.append(newData[newData['physical_transport'] == stat]['weight'])
print(allStatuses)

plt.figure(figsize=(15, 6))
plt.boxplot(filtered, labels=allStatuses)
plt.show()
```
