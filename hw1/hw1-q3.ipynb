```python
print("--- 3c ---")
male = data[data['sex'] == "M"]
male = male['hr_physical']

female = data[data['sex'] == "F"]
female = female['hr_physical']

# display(male,female)

plt.figure(figsize=(15, 6))
plt.title("Hours of Physical Activity")
plt.ylabel("Hours of Activity per week")
plt.boxplot([male,female], labels=['Male','Female'])
plt.show()
```

```python
print("--- 3e ---")
comfort = data['seat_comfort']
#have to manually count each one...
counted = []
for label in comfort.unique():
    print(label)
    l = len(data[data['seat_comfort'] == label])
    counted.append(l)
    print(l)

plt.xlabel("Rating")
plt.ylabel('Counts')
plt.title('Counts of Seat Comfort Ratings')
plt.bar(comfort.unique(),counted)
```

```python
print("--- 3f ---")

hro = data['hr_online']
hrr = data['hr_reading']
plt.scatter(hro,hrr)

plt.xlabel("hrs online")
plt.ylabel('hrs reading')
plt.title('hrs online vs hrs reading')
```

```python
print("--- 3g ---")

plt.hist(hro)

plt.xlabel("hrs online")
plt.ylabel('counts')
plt.title('hrs online')
plt.show()

print("--- 3h ---")
print("it is very left skewed")
```

```python
print("--- 3i ---")

numCreds = data['credits']
plt.hist(numCreds)
print(numCreds.unique())
print("mean: ",np.mean(numCreds))
print("median: ", np.median(numCreds))
print("mode: ", (numCreds.mode()))
```
